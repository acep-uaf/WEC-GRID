```julia
using Plots
using PowerDynamics
using TimeSeries

# Helper function to convert angular speed deviation to frequency
function angular_speed_to_frequency(ω, nominal_frequency=60)
    return nominal_frequency + ω / (2 * π)
end

# Modified plotting function using PowerGridSolution's tspan method
function plot_frequency(sol::PowerGridSolution, powergrid::PowerGrid, start_time=0.0, end_time=10.0, tres=10000)
    # Find generator indices
    generator_indices = findall(bus -> typeof(bus) == FourthOrderEq, powergrid.nodes)
    labels = ["Bus $(bus)" for bus in generator_indices]  # Assuming bus names are integers

    # Get the time vector with the desired resolution using the PowerGridSolution's tspan method
    #time_vector = tspan(sol, tres)
    #time_vector = sol.t
    time_vector = LinRange(start_time, end_time, tres)

    # Initialize an empty plot
    pl = plot()

    # For each generator index, plot the frequency
    for (i, gen_index) in enumerate(generator_indices)
        # Get angular speed deviation
        ω = sol(time_vector, gen_index, :ω)
        # Convert to frequency
        frequency = angular_speed_to_frequency.(ω)
        # Plot
        plot!(pl, time_vector, frequency, label=labels[i])
    end

    # Set labels
    xlabel!(pl, "Time [s]")
    ylabel!(pl, "Frequency [Hz]")
    
    ylims!(pl, 59.94, 60.04)
    return pl
end
```

```julia
using DataFrames, Combinatorics

function perform_n_k_contingency_analysis(network_data, k; branch_only=1)
    # Initialize DataFrame with specific column types
    contingency_df = DataFrame(Line_Info = String[], Result = String[], Iterations = Int[])

    solver = optimizer_with_attributes(Ipopt.Optimizer, "print_level" => 0)

    # Filter branches based on the specified criteria
    filtered_branches = branch_only == 1 ? 
        [branch_id for (branch_id, branch_data) in network_data["branch"] if "source_id" in keys(branch_data) && branch_data["source_id"][1] == "branch"] : 
        keys(network_data["branch"])

    unique_combinations = collect(combinations(filtered_branches, k))

    for combo in unique_combinations
        network_data_n_k = deepcopy(network_data)

        # Remove selected branches for this contingency case
        for branch_id in combo
            delete!(network_data_n_k["branch"], branch_id)
        end

        # Perform the power flow analysis
        result_dict = solve_pf(network_data_n_k, ACPPowerModel, solver)
        result_status = string(result_dict["termination_status"])

        # Collect line information into a single string
        line_info = join(["($(network_data["branch"][branch_id]["f_bus"]), $(network_data["branch"][branch_id]["t_bus"]))" for branch_id in combo], ", ")

        # Insert the collected data into the DataFrame
        new_row = (Line_Info = line_info, Result = result_status, Iterations = -1)  # Using a NamedTuple for clarity
        push!(contingency_df, new_row)
    end

    return contingency_df
end
```

```julia
# N-1 Contingency Analysis Function
function perform_n_1_contingency_analysis(network_data, lines, solver)
    n_1_df = DataFrame(Line = String[], Result = Bool[])

    for line in lines
        # Create a copy of the network data for modification
        network_data_n_1 = deepcopy(network_data)

        # Simulate outage of the line
        delete!(network_data_n_1["branch"], line)

        # Solve the power flow problem
        result = solve_opf(network_data_n_1, ACPPowerModel, solver)

        # Check for convergence
        converged = result["termination_status"] == LOCALLY_SOLVED

        # Append the results to the DataFrame
        push!(n_1_df, (Line = string(line), Result = converged))
    end

    return n_1_df
end
```

```julia
using DataFrames
using Ipopt

# Function to perform N-2 contingency analysis using a precomputed DataFrame of line pairs
function perform_n_2_contingency_analysis(network_data, contingency_df)
    # Prepare the solver
    solver = optimizer_with_attributes(Ipopt.Optimizer, "print_level" => 0)

    # DataFrame to store updated results
    updated_contingency_df = DataFrame(Line1 = String[], Line2 = String[], Result = Bool[], Iterations = Int[])

    # Loop through each pair in the contingency_df
    for row in eachrow(contingency_df)
        line1 = row.Line1
        line2 = row.Line2

        # Create a copy of the network data for modification
        network_data_n_2 = deepcopy(network_data)

        # Simulate outage of both lines
        delete!(network_data_n_2["branch"], line1)
        delete!(network_data_n_2["branch"], line2)

        # Solve the power flow problem
        result = solve_opf(network_data_n_2, ACPPowerModel, solver)

        # Check for convergence
        converged = result["termination_status"] == LOCALLY_SOLVED
        itr = -1  # Replace with actual iteration count if available

        # Append the results to the updated DataFrame
        push!(updated_contingency_df, (Line1 = string(line1), Line2 = string(line2), Result = converged, Iterations = itr))
    end

    return updated_contingency_df
end
```

```julia
using PowerModels
using Plots

function visualize_power_grid(net)
    # Extract node and edge data from the PowerModels data structure
    buses = [(b["bus_i"], b["x"], b["y"]) for (k, b) in net["bus"]]
    lines = [(l["f_bus"], l["t_bus"]) for (k, l) in net["branch"]]
    
    # Create a plot
    plot()

    # Plot the lines (edges)
    for line in lines
        f_bus, t_bus = line
        f_bus_x, f_bus_y = buses[f_bus][2:3]
        t_bus_x, t_bus_y = buses[t_bus][2:3]
        plot!([f_bus_x, t_bus_x], [f_bus_y, t_bus_y], line = (:black, 2), legend = false)
    end

    # Plot the buses (nodes)
    for bus in buses
        scatter!([bus[2]], [bus[3]], label = "Bus $(bus[1])")
    end

    # Customize and display the plot
    xlabel!("X Coordinate")
    ylabel!("Y Coordinate")
    title!("Power Grid Visualization")
    display(plot)
end

# Example usage with a PowerModels test case
net =  PowerModels.parse_file("../models/grid_models/case30.m", import_all=true)
visualize_power_grid(net)
```

```julia
# N-1 Contingency Analysis Function
function perform_n_1_contingency_analysis(network_data, lines, solver)
    n_1_df = DataFrame(Line = String[], Result = Bool[])

    for line in lines
        # Create a copy of the network data for modification
        network_data_n_1 = deepcopy(network_data)

        # Simulate outage of the line
        delete!(network_data_n_1["branch"], line)

        # Solve the power flow problem
        result = solve_opf(network_data_n_1, ACPPowerModel, solver)

        # Check for convergence
        converged = result["termination_status"] == LOCALLY_SOLVED

        # Append the results to the DataFrame
        push!(n_1_df, (Line = string(line), Result = converged))
    end

    return n_1_df
end
```

```julia
using LightGraphs, GraphPlot, Plots

function visualize_grid(network_data)
    # Extract buses and branches
    buses = collect(keys(network_data["bus"]))
    branches = values(network_data["branch"])

    # Create a graph
    G = LightGraphs.SimpleGraph(length(buses))
    bus_to_index = Dict(bus => idx for (idx, bus) in enumerate(buses))

    # Add edges for each branch
    for branch in branches
        from_bus_id = branch["f_bus"]
        to_bus_id = branch["t_bus"]
        if haskey(bus_to_index, from_bus_id) && haskey(bus_to_index, to_bus_id)
            from_bus = bus_to_index[from_bus_id]
            to_bus = bus_to_index[to_bus_id]
            add_edge!(G, from_bus, to_bus)
        end
    end

    # Generate random positions for each node (bus)
    positions = Dict(i => (rand(), rand()) for i in vertices(G))

    # Draw the graph
    gplot(G, nodelabel=[buses[i] for i in vertices(G)], nodecolor=:lightblue, layout=circular_layout)
end

# Call the function with your network data
visualize_grid(network_data)
```

```julia
using Graphs, GraphPlot, Plots

function visualize_contingency(network_edges, df, index, positions)
    # Retrieve contingency result from DataFrame
    line1 = (df[index, :Line1][1], df[index, :Line1][2])
    line2 = (df[index, :Line2][1], df[index, :Line2][2])
    result = df[index, :Result]
    iterations = df[index, :Iterations]

    # Create a graph from network edges
    G = SimpleGraph()
    for edge in network_edges
        add_edge!(G, edge[1], edge[2])
    end

    # Define colors for edges
    edge_colors = [((edge[1], edge[2]) == line1 || (edge[1], edge[2]) == line2) ? :red : :black for edge in edges(G)]

    # Draw the network with fixed positions
    gplot(G, x=positions[:,1], y=positions[:,2], nodelabel=1:nv(G), nodecolor=:lightblue, edgecolor=edge_colors)

    # Annotate with detailed convergence status, iteration count, and outaged lines
    status_message = result ? "Converged" : "Not Converged"
    outage_info = "Outaged Lines: $(line1) and $(line2)"
    annotate!(positions[line1[1],1], positions[line1[1],2], text(outage_info, :left, 10))
    annotate!(positions[line2[1],1], positions[line2[1],2], text("Status: $(status_message)\nIterations: $(iterations)", :right, 10))

    plot!()
end
```
