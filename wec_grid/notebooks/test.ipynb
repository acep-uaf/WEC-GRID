```python
import matplotlib.pyplot as plt

# Assuming you have the data in a Pandas DataFrame
data = WecGrid.wec_list[0].dataframe

# Extract 'time', 'pg', and 'pt' columns
time = data["time"]
pg = data["pg"]
pt = data["pt"]

# Set the desired time range (e.g., from 300 to 43050)
start_time = 300
end_time = 43050

# Filter the data within the specified time range
filtered_data = data[(data["time"] >= start_time) & (data["time"] <= end_time)]

# Extract 'time', 'pg', and 'pt' from the filtered data
time_range = filtered_data["time"]
pg_range = filtered_data["pg"]
pt_range = filtered_data["pt"]

# Convert time values to hours
time_range_hours = time_range / 3600

# Create a larger rectangular figure with a specified size
plt.figure(figsize=(12, 8))  # Adjust the width and height as needed

# Create points for 'pg'
plt.scatter(time_range_hours, pg_range, c="blue", marker="o", label="pg")

# # Create points for 'pt'
# plt.scatter(time_range_hours, pt_range, c='red', marker='x', label='pt')

# Connect points with lines for 'pg'
plt.plot(time_range_hours, pg_range, linestyle="-", color="blue")

# # Connect points with lines for 'pt'
# plt.plot(time_range_hours, pt_range, linestyle='-', color='red')

# Label the axes and add a legend
plt.xlabel("Time (hours)")
plt.ylabel("kW")
plt.legend()

# Set the x-axis (time) range in hours
plt.xlim(start_time / 3600, end_time / 3600)

# Show the plot
plt.show()
```
